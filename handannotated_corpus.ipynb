```python
#calculate agreement score
#agreement scores
from nltk import agreement
taskdata=[[0,str(i),str(mama_label[i])] for i in range(0,len(mama_label))]+[[1,str(i),str(basil_label[i])] for i in range(0,len(basil_label))]+[[2,str(i),str(luca_label[i])] for i in range(0,len(luca_label))]+[[3,str(i),str(alex_label[i])] for i in range(0,len(alex_label))]+[[4,str(i),str(sara_label[i])] for i in range(0,len(sara_label))]+[[5,str(i),str(abdullah_label[i])] for i in range(0,len(abdullah_label))]
ratingtask = agreement.AnnotationTask(data=taskdata)
print("alpha " +str(ratingtask.alpha()))

#get only agreements as corpus
agreements = []
for a, b, c, d, e, f in zip(basil_tupl_list, luca_tupl_list, mama_tupl_list, alex_tupl_list, abdullah_tupl_list, sara_tupl_list):
    if a == b and b == c and c == d and d == e and e ==f:
        agreements.append(a)
```
